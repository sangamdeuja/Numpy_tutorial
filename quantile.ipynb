### Explain in your own words what is standard error of the mean



Standard error of the mean is also known as standard deviation of the sampling distribution. The sampling distribution always follows the normal distribution and has mean equal(near) to population and the error 1/square root of sampling size times standard devialtion of the population.

```python
#my_array=np.random.normal(5,10,10000)
my_array=np.array([12,56,34,7,90,44,90,0,9,8,7,6,545,7,0,200,5,43,4,6,9,98,8,6,323,5,545,6,8,21,32,0,3,9,3])
print(np.mean(my_array))
print(np.std(my_array))
means=[]
stds=[]

for item in range(0,100):
    
    data=np.random.choice(my_array,replace=False,size=16)
    a=np.mean(data)
    #print(np.std(data))
    stds.append(np.std(data))
    means.append(a)
    
b=np.array(means)
print(np.mean(b))
print(np.std(b))

```

```python

sm.qqplot(my_array,loc=64.25,scale=134.18,line="45")
pylab.show()
```

The above graph shows that the my_data is not from the normal distribution


```python
sm.qqplot(b,loc=np.mean(b),scale=np.std(b),line="45")
pylab.show()
```
