### FREQUENCY TABLE OR CONTINGENCY TABLE

Frequncy table(cross tabulation) is the two way frequency table representing categorical data. The totals of rows and the columns are called marginal totals. The bottom right total represents the grand total.


### PLOTTING UNIVARATE AND BIVARATE DISTRIBUTION

The first sensible thing to know about the data is know about the distribution of variables by plotting. 

Univarate distribution can be displayed from histogram, dot plot, box plot etc.

Bivarate distribution can be displayed from scatterplot, hexplot, stacked plot etc.

### VISUALIZING PAIRWISE RELATIONSHIPS IN A DATASET

Pair plot or scatter plot can be used to visualize pairwise relationships.

```python
import numpy as np
import pandas as pd
import seaborn as sns
train = pd.read_csv('train.csv')
```

```python
train.head(5)
```

```python
##Plot the count distribution (Bernoulli) of survival  
sns.countplot(data=train, x = 'Survived')
```

```python
## Separate this distribution now also by Gender (sns.catplot). Experiment with kind and hue arguments
sns.catplot(x="Sex", hue='Survived' ,kind="count", data=train)
```

```python
## Plot SibSp vs Survived per gender and Pclass using sns.catplot
sns.factorplot('SibSp', 'Survived', hue='Sex', data=train)
```

```python
sns.factorplot('SibSp', 'Survived', hue='Pclass', data=train)
```

```python
## Plot the distribution of fare for each Pclass, color by Survived, use sns.boxplot
sns.catplot(x="Pclass", y ='Fare', hue='Survived' , data=train)
```

```python
sns.boxplot(x="Pclass", y ='Fare', hue='Survived' , data=train)
```
